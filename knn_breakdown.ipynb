```python
import numpy as np
from math import sqrt
import matplotlib.pyplot as plt
import warnings
from matplotlib import style
from collections import Counter

style.use('fivethirtyeight')
# the points are called features.  2D features.  
# Features that correspond to the class of k and features that correspond to class r
dataset = {'k': [[1,2], [2,3], [3,1]], 'r':[[6,5],[7,7],[8,6]]}
new_features = [5,7]


# ii is the same as using j
for i in dataset:
  for ii in dataset[i]:
    plt.scatter(ii[0], ii[1], s=100, color=i) # for ii in daataset[i]] for i in dataset]
```

```python
def k_nearest_neighbors(data, predict, k=3):

  # length of a dictionary is equal to its number of keys
  if len(data) >= k:
    warnings.warn('K is set to a value less than total voting groups!')

  distances = []

  for group in data:

    for features in data[group]:

      # First Version
      # euclidean_distance = sqrt( (features[0] -predict[0])**2 + (features[1] - predict[1])**2 )

      # Second Version
      # euclidean_distance = np.sqrt(np.sum((np.array(features)-np.array(predict))**2))

      # Third Version - fastest
      euclidean_distance = np.linalg.norm(np.array(features)-np.array(predict))
      distances.append([euclidean_distance, group])

      votes = [i[1] for i in sorted(distances)[:k]]
      print(Counter(votes).most_common(1))
      vote_result = Counter(votes).most_common(1)[0][0]

  
  return vote_result

result = k_nearest_neighbors(dataset, new_features, k=4)
print(result)



[[plt.scatter(ii[0], ii[1], s=100, color=i) for ii in dataset[i]] for i in dataset]
# Setting color=result will turn the dot the color of the class it belongs to
plt.scatter(new_features[0], new_features[1], s=100, color=result)
plt.show()
```
