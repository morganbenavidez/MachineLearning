```python
# 18  & 19

import numpy as np
from math import sqrt
import warnings
from collections import Counter
import pandas as pd
import random
```

```python
def k_nearest_neighbors(data, predict, k=3):

  # length of a dictionary is equal to its number of keys
  if len(data) >= k:
    warnings.warn('K is set to a value less than total voting groups!')

  distances = []

  for group in data:

    for features in data[group]:

      # First Version
      # euclidean_distance = sqrt( (features[0] -predict[0])**2 + (features[1] - predict[1])**2 )

      # Second Version
      # euclidean_distance = np.sqrt(np.sum((np.array(features)-np.array(predict))**2))

      # Third Version - fastest
      euclidean_distance = np.linalg.norm(np.array(features)-np.array(predict))
      distances.append([euclidean_distance, group])

      votes = [i[1] for i in sorted(distances)[:k]]
      #print(Counter(votes).most_common(1))
      vote_result = Counter(votes).most_common(1)[0][0]
      confidence = Counter(votes).most_common(1)[0][1] / k
      #print(vote_result, confidence)

  
  return vote_result, confidence

accuracies = []

for i in range(25):


    # Upload csv to github and 
    url = 'https://raw.githubusercontent.com/morganbenavidez/Datasets/main/breast-cancer-wisconsin.data'

    # Create dataframe
    df = pd.read_csv(url)

    df.replace('?', -99999, inplace=True)

    df.drop(['id'], 1, inplace=True)

    # Convert all of the data to float type, and this is a list of lists
    full_data= df.astype(float).values.tolist()


    # Shuffle Data

    print(full_data[:10])
    print(20*'#')

    random.shuffle(full_data)

    print(full_data[:10])

    test_size = 0.2
    train_set = {2:[], 4:[]}
    test_set = {2:[], 4:[]}
    # Everything up to 20%
    train_data = full_data[:-int(test_size*len(full_data))]
    # Last 20% of the data 
    test_data = full_data[-int(test_size*len(full_data)):]


    for i in train_data:
      # -1 is referring to the last column in the data set which is the class in this case, either a 2 or a 4
      train_set[i[-1]].append(i[:-1])

    for i in test_data:
      # -1 is referring to the last column in the data set which is the class in this case, either a 2 or a 4
      test_set[i[-1]].append(i[:-1])


    correct = 0
    total = 0

    for group in test_set:

      for data in test_set[group]:

        vote, confidence = k_nearest_neighbors(train_set, data, k=5)

        if group==vote:
          correct+=1
        
        # This will print anything where the confidence is less than perfect
        else:
          print(confidence)

        total += 1

    print('Accuracy: ', correct/total)
    accuracies.append(correct/total)

print(sum(accuracies)/len(accuracies))
```
